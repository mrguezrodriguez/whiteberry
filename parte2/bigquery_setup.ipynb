```python
from google.cloud import bigquery
from google.oauth2 import service_account
import os
from dotenv import load_dotenv

load_dotenv(dotenv_path=r"c:\Users\User\Documents\GitHub\MAR26\Bootcamp Data Online\proyectos\whiteberry\.env")

# Configuración
PROJECT_ID = os.getenv("GCP_PROJECT_ID")
DATASET_ID = os.getenv("BQ_DATASET_ID")
CREDENTIALS_PATH = os.path.join(os.path.dirname(os.path.abspath("__file__")), "..", os.getenv("GOOGLE_APPLICATION_CREDENTIALS").lstrip("./"))  ## PORQUE NO ESTÁ AL MISMO NIVEL QUE .ENV

# Cliente autenticado
credentials = service_account.Credentials.from_service_account_file(
    CREDENTIALS_PATH
)
client = bigquery.Client(
    project=PROJECT_ID,
    credentials=credentials
)

# Crear dataset
dataset_ref = bigquery.Dataset(f"{PROJECT_ID}.{DATASET_ID}")
dataset_ref.location = "EU"  # Datos en Europa
dataset = client.create_dataset(dataset_ref, exists_ok=True)
print(f"Dataset {DATASET_ID} creado")
```
