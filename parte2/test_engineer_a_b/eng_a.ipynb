```python
# ---------------------------------------
# INGENIERO A - DATA MASTER
# ---------------------------------------

import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime

fake = Faker()
np.random.seed(42)

# ---------------------------------------
# 1. CATEGORIES
# ---------------------------------------

categories = [
    {"category_id": 1, "name": "Electronics", "description": "Electronic devices"},
    {"category_id": 2, "name": "Clothing", "description": "Apparel and fashion"},
    {"category_id": 3, "name": "Home", "description": "Home and furniture"},
    {"category_id": 4, "name": "Sports", "description": "Sports equipment"},
    {"category_id": 5, "name": "Books", "description": "Books and media"},
    {"category_id": 6, "name": "Beauty", "description": "Beauty and personal care"}
]

categories_df = pd.DataFrame(categories)


# ---------------------------------------
# 2. CUSTOMERS
# ---------------------------------------

N_CUSTOMERS = 600  # mínimo 500

countries_cities = {
    "Spain": ["Madrid", "Barcelona", "Valencia", "Seville"],
    "USA": ["New York", "Los Angeles", "Chicago"],
    "France": ["Paris", "Lyon", "Marseille"],
    "Germany": ["Berlin", "Munich", "Hamburg"]
}

channels = ["organic", "paid_ads", "social_media", "referral"]

customers = []

for i in range(1, N_CUSTOMERS + 1):
    country = np.random.choice(list(countries_cities.keys()))
    city = np.random.choice(countries_cities[country])

    customers.append({
        "customer_id": i,
        "firstname": fake.first_name(),
        "lastname": fake.last_name(),
        "email": fake.email(),
        "country": country,
        "city": city,
        "registration_date": fake.date_between(start_date="-3y", end_date="today"),
        "source_channel": np.random.choice(channels)
    })

customers_df = pd.DataFrame(customers)


# ---------------------------------------
# 3. PRODUCTS
# ---------------------------------------

N_PRODUCTS = 100  # mínimo 70

products = []

for i in range(1, N_PRODUCTS + 1):
    price = round(np.random.uniform(10, 500), 2)
    
    # coste menor que precio en la mayoría (80%)
    if np.random.rand() < 0.8:
        cost = round(price * np.random.uniform(0.4, 0.8), 2)
    else:
        cost = round(price * np.random.uniform(0.8, 1.2), 2)

    products.append({
        "product_id": i,
        "category_id": np.random.choice(categories_df["category_id"]),
        "name": fake.word().capitalize(),
        "created_at": fake.date_between(start_date="-2y", end_date="today"),
        "price": price,
        "cost": cost,
        "stock": np.random.randint(0, 200),  # nunca negativo
        "is_available": np.random.choice([True, False], p=[0.8, 0.2])
    })

products_df = pd.DataFrame(products)


# ---------------------------------------
# VALIDACIONES
# ---------------------------------------

# IDs únicos
assert customers_df["customer_id"].is_unique
assert products_df["product_id"].is_unique
assert categories_df["category_id"].is_unique

# mínimos
assert len(customers_df) >= 500
assert len(products_df) >= 70

# stock no negativo
assert (products_df["stock"] >= 0).all()

# coste menor que precio en la mayoría
assert (products_df["price"] > products_df["cost"]).mean() > 0.7


# ---------------------------------------
# EXPORTAR CSV
# ---------------------------------------

customers_df.to_csv("customers.csv", index=False)
categories_df.to_csv("categories.csv", index=False)
products_df.to_csv("products.csv", index=False)
```
